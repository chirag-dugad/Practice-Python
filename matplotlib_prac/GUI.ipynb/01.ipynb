```python
#BUTTON

def on_button_click():
    print("Button Clicked!")


root=tk.Tk()
root.title("Custom Button Example!")
button=tk.Button(root, text="Click Me!", command=on_button_click, fg="blue", bg="yellow")
button.pack(pady=20)
root.mainloop()
```

```python
#RADIO BUTTON (FORM)

def Submit_form():
    name=name_entry.get()
    password=password_entry.get()
    used_os=used_os_var.get()
    linked_os=linked_os_var.get()

    print("name:", name)
    print("Password:",password)
    print("Operating Software Used:", used_os)
    print("Operating Software Linked:", linked_os)

root=tk.Tk()
root.title("GUI Form Example")


name_label=tk.Label(root, text="Name:")
name_label.pack()
name_entry=tk.Entry(root)
name_entry.pack()


password_label=tk.Label(root, text="Password")
password_label.pack()
password_entry=tk.Entry(root, show="*")
password_entry.pack()

used_os_label=tk.Label(root, text="Operating Software Used:")
used_os_label.pack()
used_os_var=tk.StringVar()
used_os_options=["LINUX", "Windows XP", "Macintosh 8.0"]
for os in used_os_options:
    os_radio=tk.Radiobutton(root, text=os, variable=used_os_var, value=os)
    os_radio.pack()


linked_os_label=tk.Label(root, text="Operating Software Linked:")
linked_os_label.pack()
linked_os_var=tk.StringVar()
linked_os_options=["LINUX", "Windows XP", "Macintosh 8.0"]
for os in linked_os_options:
    os_radio=tk.Radiobutton(root, text=os, variable=linked_os_var, value=os)
    os_radio.pack()


submit_button=tk.Button(root, text="Submit", command=Submit_form)
submit_button.pack()

root.mainloop()
```

```python
#GRID MANAGER

window = Tk() # Create a window
window.title("Grid Manager Demo") # Set title

message = Message(window, text = "This Message widget occupies three rows and two columns")
message.grid(row = 1, column = 1,  columnspan = 2)
Label(window, text = "First Name:").grid(row = 1, column = 3)
Entry(window).grid(row = 1, column = 4,  pady = 5)
Label(window, text = "Last Name:").grid(row = 2, column = 3)
Entry(window).grid(row = 2, column = 4)
Button(window, text = "Get Name").grid(row = 3, padx = 5, pady = 5, column = 4, )

window.mainloop() # Create an event loop
```

```python
#MESSAGE BOX

import tkinter.messagebox
import tkinter.simpledialog
import tkinter.colorchooser

tkinter.messagebox.showinfo("showinfo", "This is an info msg")

tkinter.messagebox.showwarning("showwarning", "This is a warning")

tkinter.messagebox.showerror("showerror", "This is an error")
 
isYes = tkinter.messagebox.askyesno("askyesno", "Continue?")
print(isYes)

isOK = tkinter.messagebox.askokcancel("askokcancel", "OK?")
print(isOK)

isYesNoCancel = tkinter.messagebox.askyesnocancel("askyesnocancel", "Yes, No, Cancel?") 
print(isYesNoCancel)

name = tkinter.simpledialog.askstring("askstring", "Enter your name")
print(name)

age = tkinter.simpledialog.askinteger("askinteger", "Enter your age")
print(age)

weight = tkinter.simpledialog.askfloat("askfloat", "Enter your weight")
print(weight)
```
