```python
url = "https://raw.githubusercontent.com/dubey-dhananjay/Python/refs/heads/main/Codes/4%E1%B5%97%CA%B0_Data-Analysis_Python/4_Matplotlib/3-Histogram/Min_Max_Seasonal_IMD_2017.csv"

data = pd.read_csv(url , usecols= ["ANNUAL - MIN"])
data
```

```python
df = pd.DataFrame(data)
df
```

```python
minarray = np.array([df["ANNUAL - MIN"]])
minarray
```

```python
y, edges = np.histogram(minarray)

print(y)
print()
print(edges)
```

```python
mid = 0.5 * (edges[1:] + edges[:-1])
mid
```

```python
df.plot(kind='hist',y='ANNUAL - MIN')
plt.plot(mid,y,'-^')
plt.title('Annual Min Temperature plot(1901 - 2017)')
plt.xlabel('Temperature')
```
