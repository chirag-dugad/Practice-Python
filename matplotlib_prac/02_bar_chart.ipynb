```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dubey-dhananjay/Python/main/Codes/4%E1%B5%97%CA%B0_Data-Analysis_Python/4_Matplotlib/1-Line_Chart/MelaSales.csv", delimiter = ",")

df.plot(kind="bar", x="Day", title="Mela Sales Report")
plt.ylabel("Sales in Rs")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dubey-dhananjay/Python/main/Codes/4%E1%B5%97%CA%B0_Data-Analysis_Python/4_Matplotlib/1-Line_Chart/MelaSales.csv", delimiter = ",")

df.plot(kind="bar", x="Day", title="Mela Sales Report", color=["red", "yellow", "purple"], edgecolor="Green", linewidth=2, linestyle="--")
plt.ylabel("Sales in Rs")
plt.xlabel("Din")
plt.show()
```
