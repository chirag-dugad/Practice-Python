```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

```python
subjects = ["egd" , "maths" , "phy" , "beee"]
marks = [10 , 15 , 14 , 15]
```

```python
plt.plot(subjects , marks , marker = "*" , color = "green" , linestyle = "dashdot")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.title("CIE 1 MARKS")
plt.grid(True)
plt.yticks(marks)
```

```python
df = pd.DataFrame({"subjects" : ["egd" , "maths" , "phy" , "beee"], "marks" : [10 , 15 , 14 , 15]})
df.plot(kind="bar" , x="subjects" , color = "yellow" , edgecolor = "green" , linestyle = "--" , )
plt.ylabel("marks")
plt.title("CIE 1 MARKS")
```

```python
sales = ({"month" : ["jan" , "feb" , "march" , "april" , "may"] ,
         "store a" : [120,150,170,200,220] ,
         "store b" : [100,130,160,180,210]})

df = pd.DataFrame(sales)

df.plot(kind="bar" , x="month")

```

```python
expenses = ({"categories" : ["food" , "transport" , "rent" , "entertainment" , "savings"] , 
             "amount" : [2500,800,4000,1200,1500]})

df = pd.DataFrame(expenses)
df.plot(kind="pie" , y="amount" , labels = df.categories , legend=False , autopct='%1.0f%%' , startangle = 90)
plt.show()
```

```python
Height = [150,160,170,180,190] 
Weight = [50,60,70,80,90]


plt.scatter(x=Height , y = Weight , marker="*" , color = "red")
plt.xlabel("height in cm ")
plt.ylabel("weight in kg ")
plt.title("weight v heigt")
```
