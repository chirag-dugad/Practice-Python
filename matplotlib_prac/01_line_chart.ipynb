```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
date = ["27/09" , "20,11" , "1/12"]
temp = [30 , 50 , 10]
plt.plot(date,temp)

```

```python
plt.plot(date,temp)
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("Date Wise Temperature")
plt.grid(True)
```

```python
plt.plot(date, temp)
plt.xlabel("Date")                      # add the Label on x-axis
plt.ylabel("Temperature")               # add the Label on y-axis
plt.title("Date wise Temperature")      # add the Title to the chart
plt.grid(True)                          # add gridlines to the background
plt.yticks(temp)                        # add the exact numbers required on chart 
```

```python
marks = [15,14,15,10]
subject = ["maths" , "beee" , "phy" , "egd"]

plt.plot(subject,marks)
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.title("Marks for CIE-1")
plt.grid(True)
plt.yticks(marks)
```

```python
height=[121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]
df=pd.DataFrame({"height":height,"weight":weight})
df
```

```python
plt.xlabel("Weight in kg")
plt.ylabel("Height in cm")
plt.title("Average weight with respect to average height")
plt.plot(df.weight , df.height , marker = "*" , markersize=18 , color = "yellow" , linewidth=2 )
```

```python
plt.xlabel('Weight in kg')
plt.ylabel('Height in cm')
plt.title('Average weight with respect to average height')
plt.plot(df.weight,df.height,marker='*',markersize=10,color='green',linewidth=1, linestyle='dashdot')
```

```python
height=[121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]
df=pd.DataFrame({"height":height,"weight":weight})
df

plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Height v Weight")
plt.plot(df.weight , df.height , color="red" , linewidth=1 , linestyle="dashdot" ,marker="*" , markersize=9)
```

```python
Week1 = [5000, 5900, 6500, 3500, 4000, 5300, 7900]
Week2 = [4000, 3000, 5000, 5500, 3000, 4300, 5900]
Week3 = [4000, 5800, 3500, 2500, 3000, 5300, 6000]

MelaSales = pd.DataFrame({"Week 1": Week1, "Week 2": Week2, "Week 3": Week3})
MelaSales
```

```python
MelaSales.plot( kind = "line" ,color = ["red", "blue", "brown"] , marker="*" , linestyle="--")
plt.title("Mela Sales Report")
plt.xlabel("Days")
plt.ylabel("Sales in Rs")
```
